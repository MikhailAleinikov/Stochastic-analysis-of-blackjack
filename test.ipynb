```python
from blackjack import Card, CardValue, Suit, Game, Hand, Moves, GameState
game = Game(5, vocal=True)
game.takeBets()
game.dealInitialCards()
```

```python
def Test(hand):
    a = input()
    if a == 'hit': return Moves.HIT
    if a == 'stand': return Moves.STAND
    if a == 'double': return Moves.DOUBLE
    if a == 'split': return Moves.SPLIT
    return Test(hand)
```

```python
while game.game_state == GameState.PLAYERS:
    game.round(Test)
```
